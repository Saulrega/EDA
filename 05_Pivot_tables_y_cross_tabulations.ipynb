# Llamar los datos 

```
#Llamamos el set de datos publico 
Chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'

```

```
import pandas as pd 
```

```
Chicago_df=pd.read_csv(Chicago_data)
```

```
Chicago_df.head()
```

*Transformaciones * básicas  

```
Chicago_df['CRASH_DATE'] = Chicago_df['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce', utc=True))
```

```
Chicago_df.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'BEAT_OF_OCCURRENCE':'sum',})
```

```
table = pd.pivot_table(Chicago_df, index =["LIGHTING_CONDITION","REPORT_TYPE"])
```

```
 Chicago_df.filter(["REPORT_TYPE","CRASH_DAY_OF_WEEK"],axis=1)
```

```
cross_table=pd.crosstab(Chicago_df["REPORT_TYPE"], Chicago_df["REPORT_TYPE"])
```

```
pd.crosstab(Chicago_df["TRAFFIC_CONTROL_DEVICE"],
            Chicago_df["WEATHER_CONDITION"],
            margins=True,
            margins_name="LIGHTING_CONDITION")
```

```
pd.crosstab(Chicago_df["TRAFFIC_CONTROL_DEVICE"], Chicago_df["WEATHER_CONDITION"],normalize=True).head(10)
```

```
pd.melt(Chicago_df, id_vars=['TRAFFIC_CONTROL_DEVICE'], value_vars=['WEATHER_CONDITION'])
```
